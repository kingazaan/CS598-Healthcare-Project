```python
# having issues because of lowercase headers in the mimic-iii file, so i have some code
# to make all the headers into capital letters here:

import os
import csv

# Specify the path to your folder of CSV files
folder_path = 'C:/Users/azaan/OneDrive/Documents/GitHub/CS598-Healthcare-Project/MIMIC-III Clinical Database/'

# Loop through each CSV file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)

        # Open the CSV file for reading
        with open(file_path, 'r') as csv_file:
            reader = csv.reader(csv_file)

            # Create a new list to hold the capitalized header row
            new_rows = []

            # Loop through each row in the CSV file
            for i, row in enumerate(reader):

                # Capitalize the header row
                if i == 0:
                    new_row = [cell.upper() for cell in row]
                    new_rows.append(new_row)

                # Keep all other rows the same
                else:
                    new_rows.append(row)

        # Open the CSV file for writing
        with open(file_path, 'w', newline='') as csv_file:
            writer = csv.writer(csv_file)

            # Write the new rows to the CSV file
            writer.writerows(new_rows)
```
